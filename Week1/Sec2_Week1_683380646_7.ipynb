```python
#การคำนวณพื้นที่สี่เหลี่ยม
width = 5
length = 8
area = width*length
print(area)
print(type(area))
```

ได้ผลลัพธ์ : 40

ทำไมถึงเป็น type int : เพราะขนาดพื้นที่เป็นเลขจำนวนเต็มคูณกัน ผลลัพธ์จึงไม่ใช่ทศนิยม(float)

```python
#การแปลงอุณหภูมิจากเซลเซียสเป็นฟาเรนไฮต์
celsius = 40
fahrenheit = (celsius * 9/5) + 32
print("Fahrenheit is",fahrenheit)
print(type(fahrenheit))
```
