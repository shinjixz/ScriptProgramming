Activity 3.2 : Pattern Printing with Nested Loops

Part 1 : Square Pattern

```python
for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

Part 2 : Right Triangle

```python
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
```

Part 3 : Challenge - Inverted Triangle

```python
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

Part 4 : Challenge - Pyramid

```python
rows = 5

for i in range(rows):
    print(" " * (rows - i - 1), end="")
    print("*" * (2 * i + 1))
```
