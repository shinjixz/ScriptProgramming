Lab 3.2: โปรแกรมการนับถอยหลัง & เกมทายตัวเลข (Countdown & User Guessing Game)
วัตถุประสงค์: ให้นักศึกษาฝึกใช้งานลูป while สำหรับการนับถอยหลังและสร้างเกมทายตัวเลขอย่างง่าย รวมถึงการนำคำสั่ง break มาใช้เพื่อออกจากลูปก่อนกำหนด

Part 1: โปรแกรมตัวนับถอยหลัง (Countdown Timer):

```python
# --- Lab 3.2 Part 1: Countdown Timer ---
start = int(input("Enter starting number for countdown: "))

while start >= 0:
    print(f"Countdown: {start}")
    start -= 1

print("Blast off!")
```

Part 2: เกมทายตัวเลขอย่างง่าย (Simple Guessing Game):

```python
# --- Lab 3.2 Part 2: Simple Guessing Game ---
secret_number = 42

while True:
    guess = int(input("Guess the number: "))

    if guess < secret_number:
        print(f"{guess} is too low! Try again.")
    elif guess > secret_number:
        print(f"{guess} is too high! Try again.")
    else:
        print("Congratulations! You guessed it!")
        break
```
