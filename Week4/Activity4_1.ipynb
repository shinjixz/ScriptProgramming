Activity 4.1: List vs. Tuple

1. ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์

ตอบ: List
เหตุผล: ข้อมูลอุณหภูมิอาจมีการแก้ไขหรือเพิ่มข้อมูลได้ในภายหลัง จึงควรใช้ List ซึ่งสามารถเปลี่ยนแปลงข้อมูลได้ (Mutable)

2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))

ตอบ: Tuple
เหตุผล: ค่าสี RGB เป็นค่าคงที่ของสีหนึ่ง ๆ และไม่ควรถูกแก้ไข จึงเหมาะกับ Tuple ซึ่งไม่สามารถเปลี่ยนแปลงข้อมูลได้ (Immutable)

3. รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้

ตอบ: List
เหตุผล: สินค้าในรถเข็นสามารถเพิ่มหรือลบได้ตลอดเวลา จึงควรใช้ List ที่รองรับการแก้ไขข้อมูล

4. รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)

ตอบ: Tuple
เหตุผล: วันในสัปดาห์เป็นข้อมูลที่มีลำดับแน่นอนและไม่เปลี่ยนแปลง จึงเหมาะกับ Tuple

5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่าง ๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้

ตอบ: List
เหตุผล: คะแนนอาจมีการแก้ไขหรืออัปเดตภายหลัง จึงควรใช้ List ที่สามารถเปลี่ยนแปลงข้อมูลได้

6. พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่

ตอบ: Tuple
เหตุผล: พิกัด GPS ของสถานที่คงที่และไม่ควรเปลี่ยนแปลง จึงเหมาะกับ Tuple ที่เป็นข้อมูลแบบ Immutable
