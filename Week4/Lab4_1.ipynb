Lab 4.1: Student Management System (List Operations)

```python
student_names = []

while True:
    print("\n===== Student Management =====")
    print("1. Add Student")
    print("2. Remove Student")
    print("3. Find Student")
    print("4. Sort Students")
    print("5. View Students")
    print("6. Exit")

    choice = input("Choose: ")

    if choice == "1":
        student_id = input("Enter Student ID: ")
        student_name = input("Enter Student Name: ")

        student_names.append([student_id, student_name])

        print("\nCurrent Students:")
        for student in student_names:
            print("ID:", student[0], "| Name:", student[1])

    elif choice == "2":
        student_id = input("Enter Student ID to remove: ")

        found = False
        for student in student_names:
            if student[0] == student_id:
                student_names.remove(student)
                print("Removed.")
                found = True
                break

        if not found:
            print("Student not found.")

    elif choice == "3":
        student_id = input("Enter Student ID to find: ")

        found = False
        for student in student_names:
            if student[0] == student_id:
                print("Found:", student[0], "-", student[1])
                found = True
                break

        if not found:
            print("Student not found.")

    elif choice == "4":
        student_names.sort(key=lambda x: x[1])
        print("Students sorted by name.")

    elif choice == "5":
        print("\nStudent List")
        for student in student_names:
            print("ID:", student[0], "| Name:", student[1])

        print("Total:", len(student_names))

    elif choice == "6":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
