Lab 4.2: Geometric Calculations (Tuple)

```python
import math

print("Enter Point 1")
x1 = float(input("x1: "))
y1 = float(input("y1: "))

print("Enter Point 2")
x2 = float(input("x2: "))
y2 = float(input("y2: "))

point1 = (x1, y1)
point2 = (x2, y2)

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)

print("Distance =", distance)
```
