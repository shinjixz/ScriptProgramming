Activity 4.2 : Interactive To-Do List

```python
todo_list = []

while True:
    print("\n===== To-Do List =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Exit")

    choice = input("Choose: ")

    if choice == "1":
        task = input("Enter task: ")
        todo_list.append(task)
        print("Task added.")

    elif choice == "2":
        if len(todo_list) == 0:
            print("No tasks.")
        else:
            print("\nYour Tasks:")
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

    elif choice == "3":
        if len(todo_list) == 0:
            print("No tasks to remove.")
        else:
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

            index = int(input("Enter task index to complete: "))

            if 0 <= index < len(todo_list):
                completed = todo_list.pop(index)
                print(completed, "completed.")
            else:
                print("Invalid index.")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
