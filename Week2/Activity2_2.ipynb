1. แนวคิดของเรื่อง (Story Concept)

ชื่อเรื่อง: การผจญภัยในป่าลึกลับ

คุณกำลังเดินสำรวจป่าลึกลับ และพบทางแยก คุณต้องเลือกเส้นทางเดิน การตัดสินใจของคุณจะนำไปสู่จุดจบที่แตกต่างกัน

2. จุดตัดสินใจ (Decision Points)
จุดตัดสินใจที่ 1

คำถาม: คุณจะเลือกไปทางไหน?

ตัวเลือก

ซ้าย (Left)
ขวา (Right)

ผลลัพธ์

ซ้าย → พบสะพานไม้เก่า
ขวา → พบหมาป่าตัวใหญ่
จุดตัดสินใจที่ 2A (ถ้าเลือกซ้าย)

คำถาม: คุณจะทำอย่างไรกับสะพานไม้?

ตัวเลือก

เดินข้ามสะพาน
เดินกลับ

ผลลัพธ์

เดินข้ามสะพาน → พบหีบสมบัติ (ชนะเกม)
เดินกลับ → ออกจากป่าอย่างปลอดภัย
จุดตัดสินใจที่ 2B (ถ้าเลือกขวา)

คำถาม: หมาป่าขวางทางอยู่ คุณจะทำอย่างไร?

ตัวเลือก

วิ่งหนี
ยืนนิ่ง

ผลลัพธ์

วิ่งหนี → หมาป่าตามทัน (Game Over)
ยืนนิ่ง → หมาป่าเดินจากไป คุณหนีรอดได้

เริ่มเกม

แสดงข้อความ "ยินดีต้อนรับสู่ป่าลึกลับ"

รับค่า choice1

IF choice1 == "ซ้าย"
    แสดงข้อความ "คุณพบสะพานไม้เก่า"

    รับค่า choice2

    IF choice2 == "ข้าม"
        แสดงข้อความ "คุณพบหีบสมบัติ! คุณชนะ!"
    ELSE IF choice2 == "กลับ"
        แสดงข้อความ "คุณออกจากป่าอย่างปลอดภัย"
    ELSE
        แสดงข้อความ "เลือกไม่ถูกต้อง"

ELSE IF choice1 == "ขวา"
    แสดงข้อความ "คุณพบหมาป่าตัวใหญ่"

    รับค่า choice2

    IF choice2 == "วิ่ง"
        แสดงข้อความ "หมาป่าตามทัน เกมจบ"
    ELSE IF choice2 == "ยืนนิ่ง"
        แสดงข้อความ "หมาป่าเดินจากไป คุณหนีรอด!"
    ELSE
        แสดงข้อความ "เลือกไม่ถูกต้อง"

ELSE
    แสดงข้อความ "เลือกไม่ถูกต้อง"
