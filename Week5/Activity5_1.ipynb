**รูปแบบ:** กิจกรรมระดมความคิด (กลุ่มย่อย หรือ ทั้งชั้นเรียน)

## คำชี้แจง

ให้นักเรียนช่วยกันยกตัวอย่างสถานการณ์หรือปัญหาในชีวิตจริงที่ **Dictionary** และ **Set** เหมาะสมจะถูกนำมาใช้เป็นโครงสร้างข้อมูล พร้อมอธิบายเหตุผลประกอบ

## คำถามกระตุ้นความคิด

### 1. คุณจะนำ Dictionary ไปใช้ตรงไหนในเกมได้บ้าง?

**แนวคำตอบ**

ตัวอย่าง: ช่องเก็บของของผู้เล่น (Inventory)

```python
inventory = {
    "Potion": 10,
    "Sword": 1,
    "Arrow": 50
}
```

**เหตุผล**

- ใช้ชื่อไอเท็มเป็น Key
- ใช้จำนวนไอเท็มเป็น Value
- ค้นหาและแก้ไขข้อมูลได้รวดเร็ว

---

### 2. Dictionary จะช่วยจัดเก็บข้อมูลสินค้าในร้านค้าออนไลน์ได้อย่างไร?

**แนวคำตอบ**

```python
products = {
    "P001": {
        "name": "Keyboard",
        "price": 890,
        "stock": 15
    },
    "P002": {
        "name": "Mouse",
        "price": 450,
        "stock": 30
    }
}
```

**เหตุผล**

- ใช้รหัสสินค้าเป็น Key
- เก็บรายละเอียดสินค้าเป็น Value
- ค้นหาข้อมูลสินค้าได้อย่างรวดเร็ว

---

### 3. Set จะมีประโยชน์อย่างไรในการจัดการสิทธิ์ของผู้ใช้ หรือการจัดการแท็ก?

**แนวคำตอบ**

ตัวอย่างการจัดการสิทธิ์

```python
permissions = {"read", "write", "delete"}

if "write" in permissions:
    print("User can write")
```

ตัวอย่างการจัดการแท็ก

```python
hashtags = {"python", "coding", "AI", "education"}
```

**เหตุผล**

- ข้อมูลใน Set ไม่ซ้ำกัน
- ตรวจสอบการมีอยู่ของข้อมูลได้รวดเร็ว
- เหมาะสำหรับข้อมูลที่ไม่ต้องการลำดับ

---

### 4. มีสถานการณ์ไหนบ้างที่ต้องตรวจสอบอย่างรวดเร็วว่า "มีข้อมูลนี้อยู่ในกลุ่มข้อมูลขนาดใหญ่หรือไม่" โดยไม่สนใจลำดับ?

**แนวคำตอบ**

ตัวอย่าง

- รายชื่อ IP Address ที่ถูกบล็อก
- Email ที่ลงทะเบียนแล้ว
- หมายเลขนักศึกษา
- ผู้เข้าชมเว็บไซต์ (นับคนไม่ซ้ำ)

```python
blocked_ip = {
    "192.168.1.1",
    "192.168.1.5",
    "10.0.0.1"
}

if "192.168.1.5" in blocked_ip:
    print("Blocked")
```

**เหตุผล**

- Set เหมาะสำหรับข้อมูลที่ไม่ซ้ำ
- ตรวจสอบสมาชิกได้รวดเร็ว
- ไม่จำเป็นต้องเรียงลำดับข้อมูล
